```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'Downloads/customer_demographics_cleaned_v1 - Raw.csv'
customer_data = pd.read_csv(file_path)

cleaned_data = customer_data.dropna(subset=['Age', 'IncomeLevel'])

plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Customers', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(
    x='IncomeLevel',
    data=cleaned_data,
    hue='IncomeLevel', 
    dodge=False,      
    palette='pastel'   
)
plt.legend([], [], frameon=False)  
plt.title('Income Level Distribution', fontsize=16)
plt.xlabel('Income Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path_transactions = 'Downloads/customer_transactions_cleaned_v1 - Raw.csv'
transactions_data = pd.read_csv(file_path_transactions)

plt.figure(figsize=(10, 6))
sns.countplot(
    y='ProductCategory',
    data=transactions_data,
    hue='ProductCategory',  
    dodge=False,            
    palette='viridis'
)
plt.legend([], [], frameon=False)  
plt.title('Top Product Categories by Frequency', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Product Category', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

payment_method_counts = transactions_data['PaymentMethod'].value_counts()
plt.figure(figsize=(8, 8))
payment_method_counts.plot.pie(
    autopct='%1.1f%%',
    startangle=140,
    cmap='viridis',
    legend=False
)
plt.title('Payment Method Distribution', fontsize=16)
plt.ylabel('')  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path_social = 'Downloads/social_media_interactions_cleaned_v1 - Raw.csv'
social_data = pd.read_csv(file_path_social)

cleaned_social_data = social_data.dropna(subset=['Platform', 'Sentiment'])

plt.figure(figsize=(10, 6))
sns.countplot(
    x='Platform',
    data=cleaned_social_data,
    hue='Platform',  
    dodge=False,      
    palette='coolwarm'
)
plt.legend([], [], frameon=False) 
plt.title('Social Media Platform Usage', fontsize=16)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Interaction Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

interaction_sentiment = cleaned_social_data.groupby(['InteractionType', 'Sentiment']).size().unstack(fill_value=0)
interaction_sentiment.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6),
    colormap='coolwarm'
)
plt.title('Sentiments by Interaction Type', fontsize=16)
plt.xlabel('Interaction Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
